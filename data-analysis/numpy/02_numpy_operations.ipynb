```python
import numpy as np
arr = np.array([10, 20, 30, 40 , 50 , 60])     #INDEXING 
```

```python
print(arr[2])                             
```

```python
print(arr[0:2])           #SLICING         
```

```python
print(arr[:5])
```

```python
print(arr[::2])
```

```python
idx = [0, 2, 4]          #FANCY INDEXING 
print(idx)              
```

```python
mask = arr > 25  # Condition: values greater than 25
print(arr[mask])         #BOOLEAN MASKING 
```
